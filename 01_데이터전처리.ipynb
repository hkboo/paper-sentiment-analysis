```python
%%time

import pandas as pd
import re
from bs4 import BeautifulSoup
from konlpy.tag import Mecab


def _remove_whitespace(text):
  return re.sub(r'\s+', ' ', text).strip()


# 여러개 공백 한개로 변환하는 함수
def remove_newlines(text):
  text_l = _remove_whitespace(text).splitlines()
  return " ".join([text.strip() for text in text_l if text.strip()])


# HTLM 기호 제거
def remove_html_tags(text):
    #from bs4 import BeautifulSoup
    clean_text = BeautifulSoup(str(text), "lxml").text
    return clean_text

# 특수문자, 한글자모음, 이모티콘 제거
def clean_text(text):
  temp = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…《\》]', '', text) # 특수문자
  temp = re.sub('([ㄱ-ㅎㅏ-ㅣ]+)', '', temp) # 한글 자음, 한글 모음
  temp = re.sub('([A-Za-z]+)', '', temp) # 영어 제거
  temp = re.sub('([0-9]+)', '', temp) # 숫자 제거
  temp = re.sub('([♡❤✌❣♥ᆢ✊❤️✨⤵️☺️;”“]+)', '', temp) # 이모티콘 
  only_BMP_pattern = re.compile("["
                    u"\U00010000-\U0010FFFF"  #BMP characters 이외
                    "]+", flags=re.UNICODE)
  temp = only_BMP_pattern.sub(r'', temp)# BMP characters만
  emoji_pattern = re.compile("["
                        u"\U0001F600-\U0001F64F"  # emoticons
                        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                        u"\U0001F680-\U0001F6FF"  # transport & map symbols
                        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                                            "]+", flags=re.UNICODE)
  temp =  emoji_pattern.sub(r'', temp) # 유니코드로 이모티콘 지우기
  temp = _remove_whitespace(temp)
  return temp


def replace_morphs(text):
  tokens = text.split(" ")
  for i, word in enumerate(tokens):
    new_word = REPLACE_MORPHS_DICT.get(word)
    if new_word is not None:
      tokens[i] = new_word.strip()
  
  new_text = ' '.join(tokens)
  return new_text


def tokenize_text(tagger, text, target_pos=None, min_char=2):
  if target_pos is None:
    # https://openuiz.tistory.com/90
    target_pos = ['NNG', 'NNP', 'NNB', 'NP', 'VV', 'VA', 'VX', 'VV+EP', 'XR', 'VCN+EC'] 

  # tokenize a text
  tokens = tagger.pos(text)

  # filter pos
  selected_tokens = [(w, p) for w, p in tokens if p in target_pos and len(w) >= min_char]

  return selected_tokens


def filter_stopwords(tokens, stop_words):
  filtered_tokens = [(w, p) for w, p in tokens if w not in stop_words]
  return filtered_tokens


def filter_min_char(tokens, min_char):
  filtered_tokens = [(w, p) for w, p in tokens if len(w) >= min_char]
  return filtered_tokens


def add_pos(tokens):
  new_tokens = ["{}/{}".format(w, p) for w, p in tokens]
  return new_tokens


def remove_pos(tokens):
  new_tokens = [w for w, p in tokens]
  return new_tokens


def join_tokens(tokens, sign=', '):
  return sign.join(tokens)


filtered_df = input_df.copy()

## 이모티콘 제거
pattern = '[\u4e00-\u9fff]'
filtered_df['first_review'] = filtered_df['first_review'].apply(lambda x: agg(x, pattern))
filtered_df['after_review'] = filtered_df['after_review'].apply(lambda x: agg(x, pattern))
print("filtered_df 1:", filtered_df.shape)

filtered_df = filtered_df[(filtered_df['first_review'] != '') | (filtered_df['after_review'] != '')]
print("filtered_df 2:", filtered_df.shape)

stop_words_l = list(set(filtered_df['first_cate'].tolist()))
stop_words_l += list(set(filtered_df['second_cate'].tolist()))
stop_words_l += ['게장', '최초', '한달', '후기', '사용', '이용', '리뷰', '블로그',
                 '가습기', '티비', '용품', '화장품', '새우', '아기띠', '배수구',
                 '밴드', '모니터', '세제', '헬스', '홈트'] #, '이것', '저것', '그것'
stop_words_l = list(set(stop_words_l))
print("stop_words_l:", len(stop_words_l))
print(stop_words_l)
```

- 줄바꿈 기호를 공백으로 치환
- HTML TAG 제거
- 특수문자, 이모티콘, 한글자모음, 영문자, 숫자 제거
- 리뷰 ( 최초 / 한달 )
  - 형분석 및 스탑워드 제거

```python
%%time

tokenized_df = filtered_df.copy()
mecab = Mecab()

tokenized_df['new_first_review'] = tokenized_df['first_review'].apply(lambda x: remove_html_tags(remove_newlines(x)))
tokenized_df['new_after_review'] = tokenized_df['after_review'].apply(lambda x: remove_html_tags(remove_newlines(x)))
```

```python
%%time

tokenized_df['new_first_review'] = tokenized_df['new_first_review'].apply(lambda x: replace_morphs(clean_text(x)))
tokenized_df['new_after_review'] = tokenized_df['new_after_review'].apply(lambda x: replace_morphs(clean_text(x)))
```

```python
%%time

## 선 리뷰
tokenized_df['new_first_review_tokens'] = tokenized_df['new_first_review'].apply(lambda x: tokenize_text(mecab, x))
tokenized_df['new_first_review_tokens'] = tokenized_df['new_first_review_tokens'].apply(lambda x: filter_stopwords(x, stop_words_l))
```

```python
%%time

## 후 리뷰
tokenized_df['new_after_review_tokens'] = tokenized_df['new_after_review'].apply(lambda x: tokenize_text(mecab, x))
tokenized_df['new_after_review_tokens'] = tokenized_df['new_after_review_tokens'].apply(lambda x: filter_stopwords(x, stop_words_l))
print("tokenized_df 1:", tokenized_df.shape)
```

```python
%%time

tokenized_df['new_first_review_token_pos_str'] = tokenized_df['new_first_review_tokens'].apply(lambda x: join_tokens(add_pos(x))) 
tokenized_df['new_after_review_token_pos_str'] = tokenized_df['new_after_review_tokens'].apply(lambda x: join_tokens(add_pos(x)))
tokenized_df['new_first_review_tokens_str'] = tokenized_df['new_first_review_tokens'].apply(lambda x: join_tokens(remove_pos(x))) 
tokenized_df['new_after_review_tokens_str'] = tokenized_df['new_after_review_tokens'].apply(lambda x: join_tokens(remove_pos(x)))
```
